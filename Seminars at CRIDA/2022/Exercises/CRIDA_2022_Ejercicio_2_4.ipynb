```
from transformers import TFGPT2LMHeadModel, GPT2Tokenizer
from transformers import pipeline, set_seed
```

## Transformers library.
The Hugging Face transformers package is an immensely popular Python library providing pretrained models that are extraordinarily useful for a variety of natural language processing (NLP) tasks.

The library provides APIs that download the pre-trained models. Once the pre-trained models are downloaded, the high-level research on the domains of NLU and NLG can be performed easily.

## LOAD THE MODEL

```
generator = pipeline('text-generation', model='DeepESP/gpt2-spanish')

```

## GENERATE TEXT

```
set_seed(42)
generator("Hoy es el tercer día del curso,", max_length=30, num_return_sequences=5, temperature=2.5)
```
