# Explicacion de Cross Validation en time series
## Walk Fordward testing.
Dividimos nuestra serie de datos original entre dos subconjuntos Train y Test. Test son los últimos datos y no haremos medidas de error sobre ellos. Sirven de control.

## K-fold
Con los datos de train, la forma de entrenar con Cross Validation, sería dividir los datos en k subconjuntos. Para cada uno de los hiperparámetros del modelo que vayamos a aplicar (xgboost, linear regression, etc) lo entrenamos con k-1 datos y validamos con el que queda. Para cada caso medimos el error de validación y hacemos la media de los k errores. Tendremos por tanto para cada combinación de hiperparámetros un error asociado.
Finalmente seleccionamos los parametros que menor error nos han dado en cross validation y entrenamos el modelo completo.

## Croos validation en time series.
La idea es la misma, sólamente cambia que en vez de aplicar k-fold construimos los k subconjuntos de la siguiente forma. Supongamos que la idea es hacer prediccion a h=1 y tenemos en la time series de train 10 datos. Podríamos construir 9 subconjuntos ordenados: ([t1, t2], [(t1,t2), t3], [(t1,t2,t3), t4], [(t1,t2,t3,t4), t5].... Donde para cada serie [], el primer subconjunto corresponde a train y el segundo a validacion.
El procedimiento es el mismo, para cada parametros del modelo realizamos CV y nos quedamos con los parámetros que minimicen el error en validation. Posteriormente se entrena el conjunto de train con esos parámetros y se hacen predicciones.


## Procedimiento
Utilizamos la herramienta TimeSeriesSplit para generar un monton de conjuntos train-test con los que hacer crossvalidation.
Los modelos los podemos ajustar a cada uno de estos pares train-test
