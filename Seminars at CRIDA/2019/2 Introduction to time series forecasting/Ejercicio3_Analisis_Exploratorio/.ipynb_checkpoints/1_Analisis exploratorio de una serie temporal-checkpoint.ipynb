# Análisis exploratorio de una serie temporal

Los primeros pasos para realizar predicciones sobre una serie temporal son realizar un análisis exploratorio del problema y obtener métricas del mismo

```python
import pandas as pd
# Cargamos los datos 
file = "data_serie.csv"
df = pd.read_csv(file, index_col=0, header=None, parse_dates=True)
df.head()
```

```python
# Renombramos los encabecamientos
df.columns = ["data"]
df.index.name = "fecha"
df.head()
```

```python
data = df

# Vamos a convertir la fecha en una serie de columnas de features
data['Year'] = data.index.year
data['Month'] = data.index.month
data['Day'] = data.index.day
data['WeekOfYear'] = data.index.weekofyear

data = data[["data", "Year", "Month", "Day", "WeekOfYear"]]
data.tail(5)
```

## Graficar la serie

# Graficamos la estacionalidad de la serie

```python
f, (ax2,ax3) = plt.subplots(2, figsize = (12, 16))

# Influencia mensual
by_month = data.groupby("Month").mean()["data"]
by_month.index = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
by_month.plot(title = 'Monthly influence', style=['-'], lw = 4, ax= ax2)


# Influencia anual
by_year = data.groupby("Year").sum()["data"]
by_year.index = by_year.index.tolist()
by_year.plot(title = 'Year influence', style=['-'], lw = 4, ax = ax3)
plt.show()
```

## Media movil de ventanas

```python
# Tail-rolling average transform
rolling = data["data"].rolling(window=52)
rolling_mean = rolling.mean()
# plot original and transformed dataset
f, ax1 = plt.subplots(1, figsize = (12, 6))
data["data"].plot()
rolling_mean.plot(color='red', ax=ax1)
pyplot.show()
```

# Histograma de la serie

```python
f = plt.hist(data["data"], bins=40)
```

# Descomposición de la serie

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(data["data"], model='multiplicative', freq=53)
result.plot()
```

# Estudio de los valores de los residuos de la serie

```python
dummy = plt.hist(result.resid[result.resid > -10**-10], bins = 30)
```

## Matriz de correlacion features

```python
# Construimos una matriz de correlaciones
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Construimos todas las correlaciones
corr_all = data.corr()

# Mask para el triangulo superior
mask = np.zeros_like(corr_all, dtype = np.bool)
mask[np.triu_indices_from(mask)] = True

f, ax = plt.subplots(figsize = (11, 9))
sns.heatmap(corr_all, mask = mask, square = True, linewidths = .5, ax = ax, cmap = "BuPu")      
plt.show()
```
